```python
# 틀림
def reverse_num(num):
    return int(str(num)[::-1])

def answer(A, B):
    re_a = reverse_num(A)
    re_b = reverse_num(B)

    return(max(re_a, re_b))


A, B = map(int, input().split())
answer(A, B)
```

```python
# 맞음
num1, num2 = input().split()
num1 = int(num1[::-1])  # [::-1] : 역순
num2 = int(num2[::-1])

print(num1) if num1 > num2 else print(num2)
```
