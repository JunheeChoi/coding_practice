```python
N = list(map(int, input().split()))  # 여러 개의 숫자를 입력받아 리스트로 저장합니다.
N.sort(reverse=True)  # 리스트를 내림차순으로 정렬합니다.
print(N)  # 정렬된 리스트를 출력합니다.
```

```python
# 정답

nums = list(map(int, str(input())))
nums.sort(reverse=True)
for i in nums:
    print(i,end='')
```
