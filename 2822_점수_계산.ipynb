# 백준 2822: 점수 계산
image.png

```python

s_dict = {}
for i in range(8):
    scr = int(input())
    s_dict[i+1] = scr

import operator

sorted_dict = dict(sorted(s_dict.items(), key=operator.itemgetter(1), reverse=True))
selected_keys = list(sorted_dict.keys())[:5]  # Select the first 5 keys (top 5 maximum values)

sum_of_values = sum(sorted_dict[key] for key in selected_keys)
print(sum_of_values)
print(" ".join(sorted(str(key) for key in selected_keys)))
```
